```python
#### matplotlib基礎-figureやaxesでのグラフのレイアウト
#### https://qiita.com/tsuruokax/items/90167693f142ebb55a7d

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
%matplotlib inline

x = np.linspace(0, 10, 10000) #わからない1

fig = plt.figure() #わからない2
ax = plt.axes() #わからない3

plt.plot(x, np.sin(x)) #同じfigureの中に追加
plt.plot(x, np.cos(x)) #同じfigureの中に追加
```

```python
### figure1つのとき
x = np.linspace(0, 10, 20)
y1 = x
y2 = x ** 2 

# figure は 1 つ
plt.figure(figsize=(6, 2)) # figureの縦横の大きさ

plt.subplot(1,2,1) # figure内の枠の大きさとどこに配置している。subplot(行の数,列の数,何番目に配置しているか)
plt.plot(x, y1)

plt.subplot(1,2,2)
plt.plot(x, y2)
```

```python
### figures2つのとき
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y1)

# figure 2 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y2)
```

```python
### subplotsとax[x,x]
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

fig, ax = plt.subplots(2, 2, figsize=(6, 4))

# 左上
ax[0, 0].plot(x, y1)

# 右上
ax[0, 1].plot(x, y2)

# 左下
ax[1, 0].plot(x, y3)

# 右下
ax[1, 1].plot(x, y4)
```

```python
### add_subplot
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

fig = plt.figure()

# 左上
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y1)

# 右上
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y2)

# 左下
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, y3)

# 右下
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, y4)
```
